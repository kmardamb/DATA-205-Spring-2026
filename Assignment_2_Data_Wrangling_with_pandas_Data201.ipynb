## Optional Extension (Extra practice)

Create a second summary table grouped by `type` that reports: - mean
`price` - median `price` - count of listings (Use method chaining
again.)

````python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Reben80/Data201/main/Dataset/housing.csv")

df.head()
````

## PART A

````python
(
    df
    .query("price > 250000 and size > 1000")
    .assign(price_per_sqft=lambda d: d["price"] / d["size"])
    .groupby("neighborhood")
    .agg(mean_price_per_sqft=("price_per_sqft", "mean"), median_price_per_sqft=("price_per_sqft", "median"), home_count=("price_per_sqft", "size"))
    .sort_values("mean_price_per_sqft", ascending=False)
)
````

## PART B

````python
library(dplyr)

df %>%
  filter(price > 250000, size > 1000) %>%
  mutate(price_per_sqft = price / size) %>%
  group_by(neighborhood) %>%
  summarise(mean_price_per_sqft = mean(price_per_sqft), median_price_per_sqft = median(price_per_sqft), count_homes = n()) %>%
  arrange(desc(mean_price_per_sqft))
````
